```python
import cv2
import os

# ====== CONFIG ======
VIDEO_FOLDER = r"E:\Sohan\Data\N1"        # your video folder
OUTPUT_FOLDER = r"E:\Sohan\Data\frames"   # where frames will be saved
TARGET_FPS = 5                             # frames per second to extract
# ====================

def extract_frames_from_video(video_path, output_dir, target_fps=5):
    cap = cv2.VideoCapture(video_path)

    if not cap.isOpened():
        print(f"Error opening video: {video_path}")
        return

    video_fps = cap.get(cv2.CAP_PROP_FPS)

    # Calculate interval between frames
    frame_interval = int(video_fps / target_fps) if video_fps > target_fps else 1

    frame_count = 0
    saved_count = 0

    while True:
        ret, frame = cap.read()
        if not ret:
            break

        if frame_count % frame_interval == 0:
            frame_filename = os.path.join(output_dir, f"frame_{saved_count:04d}.jpg")
            cv2.imwrite(frame_filename, frame)
            saved_count += 1

        frame_count += 1

    cap.release()
    print(f"Processed: {video_path} | Saved {saved_count} frames")


def process_all_videos(video_folder, output_folder, target_fps=5):
    os.makedirs(output_folder, exist_ok=True)

    for file in os.listdir(video_folder):
        if file.endswith((".mp4", ".avi", ".mov", ".mkv")):
            video_path = os.path.join(video_folder, file)

            video_name = os.path.splitext(file)[0]
            video_output_dir = os.path.join(output_folder, video_name)

            os.makedirs(video_output_dir, exist_ok=True)

            extract_frames_from_video(video_path, video_output_dir, target_fps)


# ===== RUN =====
process_all_videos(VIDEO_FOLDER, OUTPUT_FOLDER, TARGET_FPS)
```

```python
import cv2
import os
import numpy as np
from mtcnn import MTCNN
import albumentations as A

# ================= CONFIG =================
FRAME_SIZE = (224, 224)
NUM_FRAMES = 16
# ==========================================

# --------- Initialize detector ----------
mtcnn_detector = MTCNN()

# --------- Augmentation pipeline ----------
augment = A.Compose([
    A.HorizontalFlip(p=0.5),
    A.RandomBrightnessContrast(p=0.5),
    A.Rotate(limit=10, p=0.5)
])

# ==========================================
# 1. FACE DETECTION
# ==========================================
def detect_face(frame):
    rgb_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
    faces = mtcnn_detector.detect_faces(rgb_frame)

    if faces:
        # choose the largest face
        face = max(faces, key=lambda f: f['box'][2] * f['box'][3])
        x, y, bw, bh = face['box']

        x = max(0, x)
        y = max(0, y)
        bw = max(1, bw)
        bh = max(1, bh)

        cropped = frame[y:y+bh, x:x+bw]

        if cropped.size != 0:
            return cropped

    return None

# ==========================================
# 2. FACE ALIGNMENT (simple eye-based)
# ==========================================
def align_face(face):
    gray = cv2.cvtColor(face, cv2.COLOR_BGR2GRAY)

    eye_cascade = cv2.CascadeClassifier(
        cv2.data.haarcascades + 'haarcascade_eye.xml'
    )
    eyes = eye_cascade.detectMultiScale(gray)

    if len(eyes) >= 2:
        eyes = sorted(eyes, key=lambda x: x[0])
        (x1, y1, w1, h1), (x2, y2, w2, h2) = eyes[:2]

        center1 = (x1 + w1 // 2, y1 + h1 // 2)
        center2 = (x2 + w2 // 2, y2 + h2 // 2)

        dy = center2[1] - center1[1]
        dx = center2[0] - center1[0]
        angle = np.degrees(np.arctan2(dy, dx))

        center = (face.shape[1] / 2.0, face.shape[0] / 2.0)
        M = cv2.getRotationMatrix2D(center, angle, 1.0)
        aligned = cv2.warpAffine(face, M, (face.shape[1], face.shape[0]))
        return aligned

    return face

# ==========================================
# 3. PREPROCESS FRAME
# ==========================================
def preprocess_frame(frame):
    face = detect_face(frame)

    if face is None:
        return None

    face = align_face(face)
    face = cv2.resize(face, FRAME_SIZE)
    face = face.astype(np.float32) / 255.0

    return face

# ==========================================
# 4. FRAME SAMPLING
# ==========================================
def sample_frames(frames, num_frames=16):
    if len(frames) == 0:
        return np.zeros((num_frames, FRAME_SIZE[1], FRAME_SIZE[0], 3), dtype=np.float32)

    if len(frames) >= num_frames:
        idx = np.linspace(0, len(frames) - 1, num_frames).astype(int)
        sampled = [frames[i] for i in idx]
    else:
        sampled = frames.copy()
        while len(sampled) < num_frames:
            sampled.append(sampled[-1])

    return np.array(sampled, dtype=np.float32)

# ==========================================
# 5. DATA AUGMENTATION
# ==========================================
def augment_frames(frames):
    augmented = []
    for f in frames:
        img_uint8 = (f * 255).astype(np.uint8)
        aug = augment(image=img_uint8)
        img = aug['image'].astype(np.float32) / 255.0
        augmented.append(img)
    return np.array(augmented, dtype=np.float32)

# ==========================================
# 6. MAIN VIDEO PROCESS FUNCTION
# ==========================================
def process_video(video_path):
    cap = cv2.VideoCapture(video_path)
    frames = []

    while True:
        ret, frame = cap.read()
        if not ret:
            break

        processed = preprocess_frame(frame)
        if processed is not None:
            frames.append(processed)

    cap.release()

    frames = sample_frames(frames, NUM_FRAMES)
    frames = augment_frames(frames)

    return frames  # shape: (16, 224, 224, 3)

# ==========================================
# 7. PROCESS ALL VIDEOS
# ==========================================
def process_dataset(video_folder):
    data = {}

    for file in os.listdir(video_folder):
        if file.lower().endswith((".mp4", ".avi", ".mov", ".mkv")):
            path = os.path.join(video_folder, file)
            print(f"Processing {file}...")
            frames = process_video(path)
            data[file] = frames

    return data

# ===== RUN =====
if __name__ == "__main__":
    VIDEO_FOLDER = r"E:\Sohan\Data\N1"
    dataset = process_dataset(VIDEO_FOLDER)

    if len(dataset) > 0:
        print("Done! Example shape:", list(dataset.values())[0].shape)
    else:
        print("No videos found or no faces detected.")
```

# Video Preprocess using ResNet+LSTM

```python
import cv2
import os
import numpy as np
import pandas as pd
import mediapipe as mp
from mtcnn import MTCNN
import albumentations as A

# ================= CONFIG =================
VIDEO_FOLDER = r"E:\Sohan\Data\N1"
OUTPUT_FRAMES_DIR = r"E:\Sohan\Data\processed_frames"
LABELS_PATH = r"E:\Sohan\Data\labels.csv"

FRAME_SIZE = (224, 224)
NUM_FRAMES = 16
USE_MEDIAPIPE = True
# ==========================================

os.makedirs(OUTPUT_FRAMES_DIR, exist_ok=True)

# --------- Initialize detectors ----------
mp_face = mp.solutions.face_detection.FaceDetection(model_selection=1, min_detection_confidence=0.5)
mtcnn_detector = MTCNN()

# --------- Augmentation ----------
augment = A.Compose([
    A.HorizontalFlip(p=0.5),
    A.RandomBrightnessContrast(p=0.5),
    A.Rotate(limit=10, p=0.5)
])

# ==========================================
# FACE DETECTION
# ==========================================
def detect_face(frame):
    h, w, _ = frame.shape

    if USE_MEDIAPIPE:
        results = mp_face.process(cv2.cvtColor(frame, cv2.COLOR_BGR2RGB))
        if results.detections:
            bbox = results.detections[0].location_data.relative_bounding_box
            x, y, bw, bh = int(bbox.xmin*w), int(bbox.ymin*h), int(bbox.width*w), int(bbox.height*h)
            return frame[max(0,y):y+bh, max(0,x):x+bw]
    else:
        faces = mtcnn_detector.detect_faces(frame)
        if faces:
            x, y, bw, bh = faces[0]['box']
            return frame[max(0,y):y+bh, max(0,x):x+bw]

    return None

# ==========================================
# FACE ALIGNMENT
# ==========================================
def align_face(face):
    gray = cv2.cvtColor(face, cv2.COLOR_BGR2GRAY)
    eye_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_eye.xml')
    eyes = eye_cascade.detectMultiScale(gray)

    if len(eyes) >= 2:
        eyes = sorted(eyes, key=lambda x: x[0])
        (x1, y1, w1, h1), (x2, y2, w2, h2) = eyes[:2]

        center1 = (x1 + w1//2, y1 + h1//2)
        center2 = (x2 + w2//2, y2 + h2//2)

        angle = np.degrees(np.arctan2(center2[1]-center1[1], center2[0]-center1[0]))

        center = (face.shape[1] / 2.0, face.shape[0] / 2.0)
        M = cv2.getRotationMatrix2D(center, angle, 1.0)
        return cv2.warpAffine(face, M, (face.shape[1], face.shape[0]))

    return face

# ==========================================
# PREPROCESS FRAME
# ==========================================
def preprocess_frame(frame):
    face = detect_face(frame)
    if face is None:
        return None

    face = align_face(face)
    face = cv2.resize(face, FRAME_SIZE)
    face = face.astype(np.float32) / 255.0

    return face

# ==========================================
# FRAME SAMPLING
# ==========================================
def sample_frames(frames):
    if len(frames) == 0:
        return np.zeros((NUM_FRAMES, 224, 224, 3), dtype=np.float32)

    if len(frames) >= NUM_FRAMES:
        idx = np.linspace(0, len(frames)-1, NUM_FRAMES).astype(int)
        frames = [frames[i] for i in idx]
    else:
        while len(frames) < NUM_FRAMES:
            frames.append(frames[-1])

    return np.array(frames, dtype=np.float32)

# ==========================================
# AUGMENTATION
# ==========================================
def augment_frames(frames):
    aug_frames = []
    for f in frames:
        aug = augment(image=(f*255).astype(np.uint8))
        img = aug['image'].astype(np.float32) / 255.0
        aug_frames.append(img)
    return np.array(aug_frames, dtype=np.float32)

# ==========================================
# PROCESS SINGLE VIDEO
# ==========================================
def process_video(video_path):
    cap = cv2.VideoCapture(video_path)
    frames = []

    while True:
        ret, frame = cap.read()
        if not ret:
            break

        processed = preprocess_frame(frame)
        if processed is not None:
            frames.append(processed)

    cap.release()

    frames = sample_frames(frames)
    frames = augment_frames(frames)

    return frames

# ==========================================
# PROCESS DATASET + SAVE + CREATE LABEL FILE
# ==========================================
def process_dataset(video_folder):
    rows = []

    for file in os.listdir(video_folder):
        if file.endswith((".mp4", ".avi", ".mov", ".mkv")):
            video_path = os.path.join(video_folder, file)
            base_name = os.path.splitext(file)[0]

            print(f"Processing {file}...")

            frames = process_video(video_path)

            # Save frames as .npy
            save_path = os.path.join(OUTPUT_FRAMES_DIR, f"{base_name}.npy")
            np.save(save_path, frames)

            # Add to label file (empty label for now)
            rows.append({
                "file_name": base_name,
                "label": ""
            })

    # Save labels.csv
    df = pd.DataFrame(rows)
    df.to_csv(LABELS_PATH, index=False)

    print(f"\nSaved processed frames to: {OUTPUT_FRAMES_DIR}")
    print(f"Created label file: {LABELS_PATH}")


# ===== RUN =====
if __name__ == "__main__":
    process_dataset(VIDEO_FOLDER)
```

```python
import torch
import torch.nn as nn
import torchvision.models as models

class VideoFeatureExtractor(nn.Module):
    def __init__(self, hidden_dim=256, num_layers=1, bidirectional=False, dropout=0.3):
        super(VideoFeatureExtractor, self).__init__()

        # Pretrained ResNet18 as frame-level feature extractor
        resnet = models.resnet18(weights=models.ResNet18_Weights.DEFAULT)

        # Remove final classification layer
        self.cnn = nn.Sequential(*list(resnet.children())[:-1])  
        # Output per frame: (batch, 512, 1, 1)

        self.feature_dim = 512

        # LSTM for temporal modeling
        self.lstm = nn.LSTM(
            input_size=self.feature_dim,
            hidden_size=hidden_dim,
            num_layers=num_layers,
            batch_first=True,
            bidirectional=bidirectional,
            dropout=dropout if num_layers > 1 else 0.0
        )

        lstm_output_dim = hidden_dim * 2 if bidirectional else hidden_dim

        # Final feature projection
        self.fc = nn.Linear(lstm_output_dim, hidden_dim)

    def forward(self, x):
        """
        x shape: (batch_size, num_frames, channels, height, width)
        example: (8, 16, 3, 224, 224)
        """

        batch_size, num_frames, c, h, w = x.shape

        # Merge batch and time dimensions for CNN
        x = x.view(batch_size * num_frames, c, h, w)

        # Extract frame-level CNN features
        x = self.cnn(x)                      # (batch_size*num_frames, 512, 1, 1)
        x = x.view(batch_size, num_frames, self.feature_dim)  # (batch_size, num_frames, 512)

        # LSTM over frame sequence
        lstm_out, (hidden, cell) = self.lstm(x)

        # Use last time step output
        video_features = lstm_out[:, -1, :]  # (batch_size, hidden_dim) or (batch_size, hidden_dim*2)

        # Optional projection
        video_features = self.fc(video_features)

        return video_features
```

```python
import torch
import torch.nn as nn
import torchvision.models as models


class VideoFeatureExtractor(nn.Module):
    def __init__(
        self,
        hidden_dim=256,
        num_layers=1,
        bidirectional=False,
        dropout=0.3,
        freeze_cnn=False
    ):
        super(VideoFeatureExtractor, self).__init__()

        # Pretrained ResNet18 for frame-level spatial feature extraction
        resnet = models.resnet18(weights=models.ResNet18_Weights.DEFAULT)

        # Remove final classification layer
        self.cnn = nn.Sequential(*list(resnet.children())[:-1])   # output: (B*T, 512, 1, 1)
        self.feature_dim = 512

        if freeze_cnn:
            for param in self.cnn.parameters():
                param.requires_grad = False

        # LSTM for temporal modeling across frame sequence
        self.lstm = nn.LSTM(
            input_size=self.feature_dim,
            hidden_size=hidden_dim,
            num_layers=num_layers,
            batch_first=True,
            bidirectional=bidirectional,
            dropout=dropout if num_layers > 1 else 0.0
        )

        lstm_out_dim = hidden_dim * 2 if bidirectional else hidden_dim

        # Final projected video feature
        self.fc = nn.Linear(lstm_out_dim, hidden_dim)

    def forward(self, x):
        """
        x shape: (batch_size, num_frames, channels, height, width)
        Example: (8, 16, 3, 224, 224)
        """
        batch_size, num_frames, channels, height, width = x.shape

        # Merge batch and time dimensions for CNN
        x = x.view(batch_size * num_frames, channels, height, width)

        # Frame-level features
        x = self.cnn(x)                          # (B*T, 512, 1, 1)
        x = x.view(batch_size, num_frames, self.feature_dim)  # (B, T, 512)

        # Sequence modeling
        lstm_out, _ = self.lstm(x)              # (B, T, hidden_dim)

        # Take last timestep
        video_features = lstm_out[:, -1, :]     # (B, hidden_dim) or (B, hidden_dim*2)

        # Final projection
        video_features = self.fc(video_features)

        return video_features


if __name__ == "__main__":
    model = VideoFeatureExtractor(hidden_dim=256)

    # Example input: batch=4, frames=16, channels=3, size=224x224
    x = torch.randn(4, 16, 3, 224, 224)

    features = model(x)
    print("Output shape:", features.shape)   # torch.Size([4, 256])
```

# Audio Preprocess

```python
import os
import random
import numpy as np
import librosa
import soundfile as sf

# Optional noise reduction library
# pip install noisereduce
try:
    import noisereduce as nr
    NOISE_REDUCE_AVAILABLE = True
except ImportError:
    NOISE_REDUCE_AVAILABLE = False


class AudioPreprocessor:
    def __init__(
        self,
        sample_rate=16000,
        segment_duration=3.0,
        n_mels=128,
        n_fft=1024,
        hop_length=512,
        training=True
    ):
        self.sample_rate = sample_rate
        self.segment_duration = segment_duration
        self.segment_samples = int(sample_rate * segment_duration)
        self.n_mels = n_mels
        self.n_fft = n_fft
        self.hop_length = hop_length
        self.training = training

    # 1. LOAD + RESAMPLING
    def load_audio(self, audio_path):
        audio, sr = librosa.load(audio_path, sr=self.sample_rate, mono=True)
        return audio

    # 2. NOISE REDUCTION
    def reduce_noise(self, audio):
        if NOISE_REDUCE_AVAILABLE:
            reduced = nr.reduce_noise(y=audio, sr=self.sample_rate)
            return reduced
        return audio

    # 3. AUDIO NORMALIZATION
    def normalize_audio(self, audio):
        max_val = np.max(np.abs(audio))
        if max_val > 0:
            audio = audio / max_val
        return audio

    # 4. SEGMENTATION
    def segment_audio(self, audio):
        """
        Create a fixed-length segment.
        If audio is longer, take a centered segment.
        If shorter, keep as is for pad/truncate step.
        """
        if len(audio) > self.segment_samples:
            start = (len(audio) - self.segment_samples) // 2
            audio = audio[start:start + self.segment_samples]
        return audio

    # 5. PADDING / TRUNCATION
    def pad_or_truncate(self, audio):
        if len(audio) < self.segment_samples:
            pad_width = self.segment_samples - len(audio)
            audio = np.pad(audio, (0, pad_width), mode='constant')
        elif len(audio) > self.segment_samples:
            audio = audio[:self.segment_samples]
        return audio

    # 6. DATA AUGMENTATION (TRAINING ONLY)
    def augment_audio(self, audio):
        if not self.training:
            return audio

        # Random noise injection
        if random.random() < 0.5:
            noise = np.random.randn(len(audio)) * 0.003
            audio = audio + noise

        # Random pitch shift
        if random.random() < 0.5:
            n_steps = random.uniform(-2, 2)
            audio = librosa.effects.pitch_shift(
                y=audio,
                sr=self.sample_rate,
                n_steps=n_steps
            )

        # Random time stretch
        if random.random() < 0.5:
            rate = random.uniform(0.9, 1.1)
            audio = librosa.effects.time_stretch(audio, rate=rate)

            # After stretching, restore fixed length
            audio = self.pad_or_truncate(audio)

        # Random volume scaling
        if random.random() < 0.5:
            gain = random.uniform(0.8, 1.2)
            audio = audio * gain

        # Normalize again after augmentation
        audio = self.normalize_audio(audio)
        return audio

    # 7. FEATURE EXTRACTION (MEL SPECTROGRAM)
    def extract_mel_spectrogram(self, audio):
        mel = librosa.feature.melspectrogram(
            y=audio,
            sr=self.sample_rate,
            n_fft=self.n_fft,
            hop_length=self.hop_length,
            n_mels=self.n_mels
        )
        mel_db = librosa.power_to_db(mel, ref=np.max)
        return mel_db

    # FULL PIPELINE
    def process(self, audio_path):
        audio = self.load_audio(audio_path)                 # resampling included
        audio = self.reduce_noise(audio)                    # noise reduction
        audio = self.normalize_audio(audio)                 # normalization
        audio = self.segment_audio(audio)                   # segmentation
        audio = self.pad_or_truncate(audio)                 # padding/truncation
        audio = self.augment_audio(audio)                   # training only
        mel_spec = self.extract_mel_spectrogram(audio)      # feature extraction

        return {
            "audio": audio,                 # shape: (segment_samples,)
            "mel_spectrogram": mel_spec     # shape: (n_mels, time_steps)
        }


def process_audio_folder(input_folder, output_folder, training=True):
    os.makedirs(output_folder, exist_ok=True)

    preprocessor = AudioPreprocessor(training=training)

    supported_formats = (".wav", ".mp3", ".flac", ".ogg", ".m4a")

    for file_name in os.listdir(input_folder):
        if file_name.lower().endswith(supported_formats):
            audio_path = os.path.join(input_folder, file_name)
            base_name = os.path.splitext(file_name)[0]

            result = preprocessor.process(audio_path)

            mel_spec = result["mel_spectrogram"]

            # Save mel spectrogram as .npy
            save_path = os.path.join(output_folder, f"{base_name}_mel.npy")
            np.save(save_path, mel_spec)

            print(f"Processed: {file_name} -> {save_path}, shape={mel_spec.shape}")


if __name__ == "__main__":
    INPUT_FOLDER = r"E:\Sohan\Data\audio"
    OUTPUT_FOLDER = r"E:\Sohan\Data\processed_audio"

    process_audio_folder(INPUT_FOLDER, OUTPUT_FOLDER, training=True)
```

# Audio Extract using Hybrid CNN + Wav2Vec2

```python
import torch
import torch.nn as nn
import torchaudio


class MelCNNBranch(nn.Module):
    """
    CNN branch for Mel spectrogram input
    Input shape: (batch_size, 1, n_mels, time_steps)
    Output shape: (batch_size, cnn_feature_dim)
    """
    def __init__(self, cnn_feature_dim=256):
        super(MelCNNBranch, self).__init__()

        self.features = nn.Sequential(
            nn.Conv2d(1, 32, kernel_size=3, padding=1),
            nn.BatchNorm2d(32),
            nn.ReLU(),
            nn.MaxPool2d(2),

            nn.Conv2d(32, 64, kernel_size=3, padding=1),
            nn.BatchNorm2d(64),
            nn.ReLU(),
            nn.MaxPool2d(2),

            nn.Conv2d(64, 128, kernel_size=3, padding=1),
            nn.BatchNorm2d(128),
            nn.ReLU(),
            nn.AdaptiveAvgPool2d((1, 1))
        )

        self.fc = nn.Linear(128, cnn_feature_dim)

    def forward(self, mel):
        x = self.features(mel)          # (B, 128, 1, 1)
        x = x.view(x.size(0), -1)       # (B, 128)
        x = self.fc(x)                  # (B, cnn_feature_dim)
        return x


class Wav2Vec2Branch(nn.Module):
    """
    Wav2Vec2 branch for raw waveform input
    Input shape: (batch_size, num_samples)
    Output shape: (batch_size, wav2vec_feature_dim)
    """
    def __init__(self, wav2vec_feature_dim=256, freeze_wav2vec=True):
        super(Wav2Vec2Branch, self).__init__()

        bundle = torchaudio.pipelines.WAV2VEC2_BASE
        self.wav2vec = bundle.get_model()

        if freeze_wav2vec:
            for param in self.wav2vec.parameters():
                param.requires_grad = False

        # WAV2VEC2_BASE hidden size is typically 768
        self.proj = nn.Linear(768, wav2vec_feature_dim)

    def forward(self, waveform):
        """
        waveform: (B, num_samples)
        """
        if waveform.dim() == 2:
            waveform = waveform  # OK
        elif waveform.dim() == 3 and waveform.size(1) == 1:
            waveform = waveform.squeeze(1)
        else:
            raise ValueError("waveform must have shape (B, N) or (B, 1, N)")

        with torch.set_grad_enabled(any(p.requires_grad for p in self.wav2vec.parameters())):
            features, _ = self.wav2vec.extract_features(waveform)

        # features is a list of transformer layer outputs
        # take the last layer: (B, T, 768)
        x = features[-1]

        # temporal mean pooling
        x = x.mean(dim=1)               # (B, 768)
        x = self.proj(x)                # (B, wav2vec_feature_dim)
        return x


class HybridAudioFeatureExtractor(nn.Module):
    """
    Hybrid audio feature extractor:
    - CNN branch on Mel spectrogram
    - Wav2Vec2 branch on raw waveform
    - Feature fusion by concatenation

    Inputs:
        mel:      (B, 1, n_mels, time_steps)
        waveform: (B, num_samples)

    Output:
        fused_audio_features: (B, output_dim)
    """
    def __init__(
        self,
        cnn_feature_dim=256,
        wav2vec_feature_dim=256,
        fusion_hidden_dim=256,
        freeze_wav2vec=True
    ):
        super(HybridAudioFeatureExtractor, self).__init__()

        self.mel_branch = MelCNNBranch(cnn_feature_dim=cnn_feature_dim)
        self.wav_branch = Wav2Vec2Branch(
            wav2vec_feature_dim=wav2vec_feature_dim,
            freeze_wav2vec=freeze_wav2vec
        )

        fused_dim = cnn_feature_dim + wav2vec_feature_dim

        self.fusion = nn.Sequential(
            nn.Linear(fused_dim, fusion_hidden_dim),
            nn.ReLU(),
            nn.Dropout(0.3)
        )

    def forward(self, mel, waveform):
        mel_feat = self.mel_branch(mel)         # (B, cnn_feature_dim)
        wav_feat = self.wav_branch(waveform)    # (B, wav2vec_feature_dim)

        fused = torch.cat([mel_feat, wav_feat], dim=1)
        fused = self.fusion(fused)              # (B, fusion_hidden_dim)

        return fused
```

```python
class AudioEmotionModel(nn.Module):
    def __init__(self, num_classes=7):
        super(AudioEmotionModel, self).__init__()
        self.feature_extractor = HybridAudioFeatureExtractor(
            cnn_feature_dim=256,
            wav2vec_feature_dim=256,
            fusion_hidden_dim=256,
            freeze_wav2vec=True
        )
        self.classifier = nn.Linear(256, num_classes)

    def forward(self, mel, waveform):
        features = self.feature_extractor(mel, waveform)
        logits = self.classifier(features)
        return logits
```

# Fusion 
## Video features from ResNet18 + LSTM
## Audio features from CNN + Wav2Vec2

```python
import torch
import torch.nn as nn


class MultimodalFusionModel(nn.Module):
    def __init__(
        self,
        video_feature_dim=256,
        audio_feature_dim=256,
        fusion_dim=256,
        num_classes=7,
        dropout=0.3
    ):
        super(MultimodalFusionModel, self).__init__()

        # Fusion network
        self.fusion_layer = nn.Sequential(
            nn.Linear(video_feature_dim + audio_feature_dim, fusion_dim),
            nn.ReLU(),
            nn.Dropout(dropout)
        )

        # Final classifier
        self.classifier = nn.Sequential(
            nn.Linear(fusion_dim, 128),
            nn.ReLU(),
            nn.Dropout(dropout),
            nn.Linear(128, num_classes)
        )

    def forward(self, video_features, audio_features):
        """
        video_features: (B, video_feature_dim)
        audio_features: (B, audio_feature_dim)
        """

        # Concatenate audio and video features
        fused = torch.cat([video_features, audio_features], dim=1)

        # Fusion layer
        fused_features = self.fusion_layer(fused)

        # Emotion classification
        output = self.classifier(fused_features)

        return output, fused_features
```

```python
import torch
import torch.nn as nn


class FullMultimodalEmotionModel(nn.Module):
    def __init__(
        self,
        video_extractor,
        audio_extractor,
        video_feature_dim=256,
        audio_feature_dim=256,
        fusion_dim=256,
        num_classes=7,
        dropout=0.3
    ):
        super(FullMultimodalEmotionModel, self).__init__()

        self.video_extractor = video_extractor
        self.audio_extractor = audio_extractor

        self.fusion_layer = nn.Sequential(
            nn.Linear(video_feature_dim + audio_feature_dim, fusion_dim),
            nn.ReLU(),
            nn.Dropout(dropout)
        )

        self.classifier = nn.Sequential(
            nn.Linear(fusion_dim, 128),
            nn.ReLU(),
            nn.Dropout(dropout),
            nn.Linear(128, num_classes)
        )

    def forward(self, video_input, mel_input, waveform_input):
        """
        video_input:    (B, T, C, H, W)
        mel_input:      (B, 1, n_mels, time_steps)
        waveform_input: (B, num_samples)
        """

        video_features = self.video_extractor(video_input)          # (B, 256)
        audio_features = self.audio_extractor(mel_input, waveform_input)  # (B, 256)

        fused = torch.cat([video_features, audio_features], dim=1)
        fused_features = self.fusion_layer(fused)
        output = self.classifier(fused_features)

        return output, fused_features
```

# Train freeze-then-fine-tune pipeline

```python
import torch
import torch.nn as nn
from torch.utils.data import DataLoader

# -------------------------------------------------
# 1. FREEZE / UNFREEZE HELPERS
# -------------------------------------------------
def freeze_module(module):
    for param in module.parameters():
        param.requires_grad = False


def unfreeze_module(module):
    for param in module.parameters():
        param.requires_grad = True


def unfreeze_last_resnet_block(video_extractor):
    """
    Unfreeze only the last CNN block of ResNet18 inside the video extractor.
    Assumes:
        video_extractor.cnn = nn.Sequential(*list(resnet.children())[:-1])

    In ResNet18, the last main conv block is usually layer4.
    Inside this sequential:
        index 7 -> layer4
    """
    if hasattr(video_extractor, "cnn") and len(video_extractor.cnn) > 7:
        for param in video_extractor.cnn[7].parameters():
            param.requires_grad = True


def unfreeze_top_wav2vec_layers(audio_extractor, num_layers=2):
    """
    Unfreeze only top transformer layers of wav2vec branch.
    Assumes:
        audio_extractor.wav_branch.wav2vec exists
    """
    if not hasattr(audio_extractor, "wav_branch"):
        return
    if not hasattr(audio_extractor.wav_branch, "wav2vec"):
        return

    wav2vec_model = audio_extractor.wav_branch.wav2vec

    # First freeze all
    freeze_module(wav2vec_model)

    # Try to unfreeze last few encoder transformer layers
    try:
        layers = wav2vec_model.encoder.transformer.layers
        for layer in layers[-num_layers:]:
            for param in layer.parameters():
                param.requires_grad = True
    except Exception:
        # fallback: if structure differs, do nothing extra
        pass


# -------------------------------------------------
# 2. TRAIN / VALIDATION LOOP
# -------------------------------------------------
def train_one_epoch(model, dataloader, criterion, optimizer, device):
    model.train()

    running_loss = 0.0
    correct = 0
    total = 0

    for batch in dataloader:
        video_input = batch["video"].to(device)         # (B, T, C, H, W)
        mel_input = batch["mel"].to(device)             # (B, 1, n_mels, time_steps)
        waveform_input = batch["waveform"].to(device)   # (B, num_samples)
        labels = batch["label"].to(device)              # (B,)

        optimizer.zero_grad()

        logits, _ = model(video_input, mel_input, waveform_input)
        loss = criterion(logits, labels)

        loss.backward()
        optimizer.step()

        running_loss += loss.item() * labels.size(0)

        preds = torch.argmax(logits, dim=1)
        correct += (preds == labels).sum().item()
        total += labels.size(0)

    epoch_loss = running_loss / total
    epoch_acc = correct / total

    return epoch_loss, epoch_acc


@torch.no_grad()
def validate_one_epoch(model, dataloader, criterion, device):
    model.eval()

    running_loss = 0.0
    correct = 0
    total = 0

    for batch in dataloader:
        video_input = batch["video"].to(device)
        mel_input = batch["mel"].to(device)
        waveform_input = batch["waveform"].to(device)
        labels = batch["label"].to(device)

        logits, _ = model(video_input, mel_input, waveform_input)
        loss = criterion(logits, labels)

        running_loss += loss.item() * labels.size(0)

        preds = torch.argmax(logits, dim=1)
        correct += (preds == labels).sum().item()
        total += labels.size(0)

    epoch_loss = running_loss / total
    epoch_acc = correct / total

    return epoch_loss, epoch_acc


# -------------------------------------------------
# 3. OPTIMIZER BUILDER
# -------------------------------------------------
def build_optimizer(model, lr):
    trainable_params = [p for p in model.parameters() if p.requires_grad]
    return torch.optim.Adam(trainable_params, lr=lr)


# -------------------------------------------------
# 4. FREEZE-THEN-FINE-TUNE SETUP
# -------------------------------------------------
def setup_phase_1(model):
    """
    Phase 1:
    Freeze pretrained ResNet18 CNN and Wav2Vec2.
    Train:
        - LSTM in video extractor
        - final projection in video extractor
        - Mel CNN branch
        - wav2vec projection layer
        - fusion layer
        - classifier
    """
    # Freeze video CNN backbone
    if hasattr(model.video_extractor, "cnn"):
        freeze_module(model.video_extractor.cnn)

    # Freeze wav2vec backbone only
    if hasattr(model.audio_extractor, "wav_branch") and hasattr(model.audio_extractor.wav_branch, "wav2vec"):
        freeze_module(model.audio_extractor.wav_branch.wav2vec)

    # Make sure the remaining task-specific layers are trainable
    if hasattr(model.video_extractor, "lstm"):
        unfreeze_module(model.video_extractor.lstm)
    if hasattr(model.video_extractor, "fc"):
        unfreeze_module(model.video_extractor.fc)

    if hasattr(model.audio_extractor, "mel_branch"):
        unfreeze_module(model.audio_extractor.mel_branch)
    if hasattr(model.audio_extractor, "wav_branch") and hasattr(model.audio_extractor.wav_branch, "proj"):
        unfreeze_module(model.audio_extractor.wav_branch.proj)

    if hasattr(model.audio_extractor, "fusion"):
        unfreeze_module(model.audio_extractor.fusion)

    unfreeze_module(model.fusion_layer)
    unfreeze_module(model.classifier)


def setup_phase_2(model):
    """
    Phase 2:
    Fine-tune last block of ResNet18 and top wav2vec layers
    with smaller learning rate.
    """
    # Unfreeze last CNN block
    unfreeze_last_resnet_block(model.video_extractor)

    # Unfreeze top wav2vec layers
    unfreeze_top_wav2vec_layers(model.audio_extractor, num_layers=2)

    # Keep task-specific layers trainable
    if hasattr(model.video_extractor, "lstm"):
        unfreeze_module(model.video_extractor.lstm)
    if hasattr(model.video_extractor, "fc"):
        unfreeze_module(model.video_extractor.fc)

    if hasattr(model.audio_extractor, "mel_branch"):
        unfreeze_module(model.audio_extractor.mel_branch)
    if hasattr(model.audio_extractor, "wav_branch") and hasattr(model.audio_extractor.wav_branch, "proj"):
        unfreeze_module(model.audio_extractor.wav_branch.proj)

    if hasattr(model.audio_extractor, "fusion"):
        unfreeze_module(model.audio_extractor.fusion)

    unfreeze_module(model.fusion_layer)
    unfreeze_module(model.classifier)


# -------------------------------------------------
# 5. MAIN TRAINING PIPELINE
# -------------------------------------------------
def train_freeze_then_finetune(
    model,
    train_loader,
    val_loader,
    device,
    phase1_epochs=10,
    phase2_epochs=5,
    phase1_lr=1e-4,
    phase2_lr=1e-5,
    save_path="best_multimodal_model.pth"
):
    criterion = nn.CrossEntropyLoss()

    best_val_acc = 0.0
    history = {
        "phase1_train_loss": [],
        "phase1_train_acc": [],
        "phase1_val_loss": [],
        "phase1_val_acc": [],
        "phase2_train_loss": [],
        "phase2_train_acc": [],
        "phase2_val_loss": [],
        "phase2_val_acc": [],
    }

    model.to(device)

    # -------------------------------
    # PHASE 1: freeze pretrained parts
    # -------------------------------
    print("\n===== PHASE 1: Freeze pretrained backbones =====")
    setup_phase_1(model)
    optimizer = build_optimizer(model, lr=phase1_lr)

    for epoch in range(phase1_epochs):
        train_loss, train_acc = train_one_epoch(model, train_loader, criterion, optimizer, device)
        val_loss, val_acc = validate_one_epoch(model, val_loader, criterion, device)

        history["phase1_train_loss"].append(train_loss)
        history["phase1_train_acc"].append(train_acc)
        history["phase1_val_loss"].append(val_loss)
        history["phase1_val_acc"].append(val_acc)

        print(
            f"[Phase 1][Epoch {epoch+1}/{phase1_epochs}] "
            f"Train Loss: {train_loss:.4f} | Train Acc: {train_acc:.4f} | "
            f"Val Loss: {val_loss:.4f} | Val Acc: {val_acc:.4f}"
        )

        if val_acc > best_val_acc:
            best_val_acc = val_acc
            torch.save(model.state_dict(), save_path)
            print(f"Saved best model to {save_path}")

    # -------------------------------
    # PHASE 2: fine-tune selected layers
    # -------------------------------
    print("\n===== PHASE 2: Fine-tune top pretrained layers =====")
    setup_phase_2(model)
    optimizer = build_optimizer(model, lr=phase2_lr)

    for epoch in range(phase2_epochs):
        train_loss, train_acc = train_one_epoch(model, train_loader, criterion, optimizer, device)
        val_loss, val_acc = validate_one_epoch(model, val_loader, criterion, optimizer=None, device=device) if False else validate_one_epoch(model, val_loader, criterion, device)

        history["phase2_train_loss"].append(train_loss)
        history["phase2_train_acc"].append(train_acc)
        history["phase2_val_loss"].append(val_loss)
        history["phase2_val_acc"].append(val_acc)

        print(
            f"[Phase 2][Epoch {epoch+1}/{phase2_epochs}] "
            f"Train Loss: {train_loss:.4f} | Train Acc: {train_acc:.4f} | "
            f"Val Loss: {val_loss:.4f} | Val Acc: {val_acc:.4f}"
        )

        if val_acc > best_val_acc:
            best_val_acc = val_acc
            torch.save(model.state_dict(), save_path)
            print(f"Saved best model to {save_path}")

    print(f"\nTraining finished. Best validation accuracy: {best_val_acc:.4f}")
    return history


# -------------------------------------------------
# 6. EXAMPLE DUMMY DATASET
# Replace with your real dataset
# -------------------------------------------------
class DummyMultimodalDataset(torch.utils.data.Dataset):
    def __init__(self, num_samples=100, num_classes=7):
        self.num_samples = num_samples
        self.num_classes = num_classes

    def __len__(self):
        return self.num_samples

    def __getitem__(self, idx):
        sample = {
            "video": torch.randn(16, 3, 224, 224),   # (T, C, H, W)
            "mel": torch.randn(1, 128, 94),          # (1, n_mels, time_steps)
            "waveform": torch.randn(48000),          # 3 sec @ 16kHz
            "label": torch.randint(0, self.num_classes, (1,)).item()
        }
        return sample


# -------------------------------------------------
# 7. EXAMPLE USAGE
# Assumes these classes already exist:
#    - VideoFeatureExtractor
#    - HybridAudioFeatureExtractor
#    - FullMultimodalEmotionModel
# -------------------------------------------------
if __name__ == "__main__":
    from torch.utils.data import random_split

    # Replace these imports with your actual classes/files
    # from video_model import VideoFeatureExtractor
    # from audio_model import HybridAudioFeatureExtractor
    # from fusion_model import FullMultimodalEmotionModel

    # Dummy example classes for structure check only
    class VideoFeatureExtractor(nn.Module):
        def __init__(self, hidden_dim=256):
            super().__init__()
            self.cnn = nn.Sequential(
                nn.Conv2d(3, 16, 3, padding=1),
                nn.ReLU(),
                nn.AdaptiveAvgPool2d((1, 1))
            )
            self.lstm = nn.LSTM(input_size=16, hidden_size=hidden_dim, batch_first=True)
            self.fc = nn.Linear(hidden_dim, hidden_dim)

        def forward(self, x):
            b, t, c, h, w = x.shape
            x = x.view(b * t, c, h, w)
            x = self.cnn(x).view(b, t, 16)
            x, _ = self.lstm(x)
            x = x[:, -1, :]
            return self.fc(x)

    class MelCNNBranch(nn.Module):
        def __init__(self, out_dim=256):
            super().__init__()
            self.net = nn.Sequential(
                nn.Conv2d(1, 32, 3, padding=1),
                nn.ReLU(),
                nn.AdaptiveAvgPool2d((1, 1))
            )
            self.fc = nn.Linear(32, out_dim)

        def forward(self, x):
            x = self.net(x).view(x.size(0), -1)
            return self.fc(x)

    class DummyWavBranch(nn.Module):
        def __init__(self, out_dim=256):
            super().__init__()
            self.wav2vec = nn.Linear(48000, 512)
            self.proj = nn.Linear(512, out_dim)

        def forward(self, x):
            x = self.wav2vec(x)
            return self.proj(x)

    class HybridAudioFeatureExtractor(nn.Module):
        def __init__(self, fusion_hidden_dim=256):
            super().__init__()
            self.mel_branch = MelCNNBranch(out_dim=256)
            self.wav_branch = DummyWavBranch(out_dim=256)
            self.fusion = nn.Sequential(
                nn.Linear(512, fusion_hidden_dim),
                nn.ReLU()
            )

        def forward(self, mel, waveform):
            mel_feat = self.mel_branch(mel)
            wav_feat = self.wav_branch(waveform)
            x = torch.cat([mel_feat, wav_feat], dim=1)
            return self.fusion(x)

    class FullMultimodalEmotionModel(nn.Module):
        def __init__(self, video_extractor, audio_extractor, video_feature_dim=256, audio_feature_dim=256, fusion_dim=256, num_classes=7, dropout=0.3):
            super().__init__()
            self.video_extractor = video_extractor
            self.audio_extractor = audio_extractor
            self.fusion_layer = nn.Sequential(
                nn.Linear(video_feature_dim + audio_feature_dim, fusion_dim),
                nn.ReLU(),
                nn.Dropout(dropout)
            )
            self.classifier = nn.Sequential(
                nn.Linear(fusion_dim, 128),
                nn.ReLU(),
                nn.Dropout(dropout),
                nn.Linear(128, num_classes)
            )

        def forward(self, video_input, mel_input, waveform_input):
            video_features = self.video_extractor(video_input)
            audio_features = self.audio_extractor(mel_input, waveform_input)
            fused = torch.cat([video_features, audio_features], dim=1)
            fused_features = self.fusion_layer(fused)
            output = self.classifier(fused_features)
            return output, fused_features

    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

    # Build model
    video_extractor = VideoFeatureExtractor(hidden_dim=256)
    audio_extractor = HybridAudioFeatureExtractor(fusion_hidden_dim=256)
    model = FullMultimodalEmotionModel(
        video_extractor=video_extractor,
        audio_extractor=audio_extractor,
        video_feature_dim=256,
        audio_feature_dim=256,
        fusion_dim=256,
        num_classes=7
    )

    # Build dataset
    full_dataset = DummyMultimodalDataset(num_samples=120, num_classes=7)
    train_size = int(0.8 * len(full_dataset))
    val_size = len(full_dataset) - train_size
    train_dataset, val_dataset = random_split(full_dataset, [train_size, val_size])

    train_loader = DataLoader(train_dataset, batch_size=4, shuffle=True, num_workers=0)
    val_loader = DataLoader(val_dataset, batch_size=4, shuffle=False, num_workers=0)

    # Train
    history = train_freeze_then_finetune(
        model=model,
        train_loader=train_loader,
        val_loader=val_loader,
        device=device,
        phase1_epochs=3,
        phase2_epochs=2,
        phase1_lr=1e-4,
        phase2_lr=1e-5,
        save_path="best_multimodal_model.pth"
    )
```

```python
import os
import copy
import torch
import torch.nn as nn
from collections import Counter
from torch.cuda.amp import autocast, GradScaler


# -------------------------------------------------
# 1. CLASS WEIGHT CALCULATION
# -------------------------------------------------
def compute_class_weights(dataset, num_classes):
    labels = []

    for i in range(len(dataset)):
        sample = dataset[i]
        labels.append(sample["label"])

    counter = Counter(labels)
    total = len(labels)

    class_weights = []
    for cls in range(num_classes):
        count = counter.get(cls, 1)
        weight = total / count
        class_weights.append(weight)

    class_weights = torch.tensor(class_weights, dtype=torch.float32)
    class_weights = class_weights / class_weights.sum() * num_classes
    return class_weights


# -------------------------------------------------
# 2. EARLY STOPPING
# -------------------------------------------------
class EarlyStopping:
    def __init__(self, patience=5, min_delta=0.0):
        self.patience = patience
        self.min_delta = min_delta
        self.best_score = None
        self.counter = 0
        self.early_stop = False

    def step(self, score):
        if self.best_score is None:
            self.best_score = score
            return False

        if score > self.best_score + self.min_delta:
            self.best_score = score
            self.counter = 0
        else:
            self.counter += 1

        if self.counter >= self.patience:
            self.early_stop = True

        return self.early_stop


# -------------------------------------------------
# 3. TRAIN ONE EPOCH
# -------------------------------------------------
def train_one_epoch_improved(model, dataloader, criterion, optimizer, device, scaler, max_grad_norm=1.0):
    model.train()

    running_loss = 0.0
    correct = 0
    total = 0

    for batch in dataloader:
        video_input = batch["video"].to(device)
        mel_input = batch["mel"].to(device)
        waveform_input = batch["waveform"].to(device)
        labels = batch["label"].to(device)

        optimizer.zero_grad()

        with autocast(enabled=(device.type == "cuda")):
            logits, _ = model(video_input, mel_input, waveform_input)
            loss = criterion(logits, labels)

        scaler.scale(loss).backward()

        # gradient clipping
        scaler.unscale_(optimizer)
        torch.nn.utils.clip_grad_norm_(model.parameters(), max_grad_norm)

        scaler.step(optimizer)
        scaler.update()

        running_loss += loss.item() * labels.size(0)

        preds = torch.argmax(logits, dim=1)
        correct += (preds == labels).sum().item()
        total += labels.size(0)

    epoch_loss = running_loss / total
    epoch_acc = correct / total

    return epoch_loss, epoch_acc


# -------------------------------------------------
# 4. VALIDATE ONE EPOCH
# -------------------------------------------------
@torch.no_grad()
def validate_one_epoch_improved(model, dataloader, criterion, device):
    model.eval()

    running_loss = 0.0
    correct = 0
    total = 0

    all_preds = []
    all_labels = []

    for batch in dataloader:
        video_input = batch["video"].to(device)
        mel_input = batch["mel"].to(device)
        waveform_input = batch["waveform"].to(device)
        labels = batch["label"].to(device)

        with autocast(enabled=(device.type == "cuda")):
            logits, _ = model(video_input, mel_input, waveform_input)
            loss = criterion(logits, labels)

        running_loss += loss.item() * labels.size(0)

        preds = torch.argmax(logits, dim=1)
        correct += (preds == labels).sum().item()
        total += labels.size(0)

        all_preds.extend(preds.cpu().tolist())
        all_labels.extend(labels.cpu().tolist())

    epoch_loss = running_loss / total
    epoch_acc = correct / total

    return epoch_loss, epoch_acc, all_preds, all_labels


# -------------------------------------------------
# 5. IMPROVED FREEZE-THEN-FINE-TUNE TRAINING
# -------------------------------------------------
def train_freeze_then_finetune_improved(
    model,
    train_loader,
    val_loader,
    train_dataset,
    num_classes,
    device,
    setup_phase_1_fn,
    setup_phase_2_fn,
    phase1_epochs=10,
    phase2_epochs=5,
    phase1_lr=1e-4,
    phase2_lr=1e-5,
    patience=5,
    save_path="best_multimodal_model_improved.pth"
):
    model.to(device)

    # class weights
    class_weights = compute_class_weights(train_dataset, num_classes).to(device)

    # improved loss
    criterion = nn.CrossEntropyLoss(
        weight=class_weights,
        label_smoothing=0.1
    )

    scaler = GradScaler(enabled=(device.type == "cuda"))
    best_val_acc = 0.0
    best_model_wts = copy.deepcopy(model.state_dict())

    history = {
        "phase1_train_loss": [],
        "phase1_train_acc": [],
        "phase1_val_loss": [],
        "phase1_val_acc": [],
        "phase2_train_loss": [],
        "phase2_train_acc": [],
        "phase2_val_loss": [],
        "phase2_val_acc": [],
    }

    # -------------------------------------------------
    # PHASE 1
    # -------------------------------------------------
    print("\n===== PHASE 1: Improved frozen training =====")
    setup_phase_1_fn(model)

    optimizer = torch.optim.AdamW(
        [p for p in model.parameters() if p.requires_grad],
        lr=phase1_lr,
        weight_decay=1e-4
    )

    scheduler = torch.optim.lr_scheduler.ReduceLROnPlateau(
        optimizer,
        mode="max",
        factor=0.5,
        patience=2
    )

    early_stopper = EarlyStopping(patience=patience, min_delta=0.001)

    for epoch in range(phase1_epochs):
        train_loss, train_acc = train_one_epoch_improved(
            model, train_loader, criterion, optimizer, device, scaler
        )

        val_loss, val_acc, _, _ = validate_one_epoch_improved(
            model, val_loader, criterion, device
        )

        scheduler.step(val_acc)

        history["phase1_train_loss"].append(train_loss)
        history["phase1_train_acc"].append(train_acc)
        history["phase1_val_loss"].append(val_loss)
        history["phase1_val_acc"].append(val_acc)

        print(
            f"[Phase 1][Epoch {epoch+1}/{phase1_epochs}] "
            f"Train Loss: {train_loss:.4f} | Train Acc: {train_acc:.4f} | "
            f"Val Loss: {val_loss:.4f} | Val Acc: {val_acc:.4f} | "
            f"LR: {optimizer.param_groups[0]['lr']:.6f}"
        )

        if val_acc > best_val_acc:
            best_val_acc = val_acc
            best_model_wts = copy.deepcopy(model.state_dict())
            torch.save(model.state_dict(), save_path)
            print(f"Saved best model to {save_path}")

        if early_stopper.step(val_acc):
            print("Early stopping triggered in Phase 1")
            break

    # -------------------------------------------------
    # PHASE 2
    # -------------------------------------------------
    print("\n===== PHASE 2: Improved fine-tuning =====")
    model.load_state_dict(best_model_wts)
    setup_phase_2_fn(model)

    optimizer = torch.optim.AdamW(
        [p for p in model.parameters() if p.requires_grad],
        lr=phase2_lr,
        weight_decay=1e-5
    )

    scheduler = torch.optim.lr_scheduler.CosineAnnealingLR(
        optimizer,
        T_max=phase2_epochs
    )

    early_stopper = EarlyStopping(patience=patience, min_delta=0.001)

    for epoch in range(phase2_epochs):
        train_loss, train_acc = train_one_epoch_improved(
            model, train_loader, criterion, optimizer, device, scaler
        )

        val_loss, val_acc, _, _ = validate_one_epoch_improved(
            model, val_loader, criterion, device
        )

        scheduler.step()

        history["phase2_train_loss"].append(train_loss)
        history["phase2_train_acc"].append(train_acc)
        history["phase2_val_loss"].append(val_loss)
        history["phase2_val_acc"].append(val_acc)

        print(
            f"[Phase 2][Epoch {epoch+1}/{phase2_epochs}] "
            f"Train Loss: {train_loss:.4f} | Train Acc: {train_acc:.4f} | "
            f"Val Loss: {val_loss:.4f} | Val Acc: {val_acc:.4f} | "
            f"LR: {optimizer.param_groups[0]['lr']:.6f}"
        )

        if val_acc > best_val_acc:
            best_val_acc = val_acc
            best_model_wts = copy.deepcopy(model.state_dict())
            torch.save(model.state_dict(), save_path)
            print(f"Saved best model to {save_path}")

        if early_stopper.step(val_acc):
            print("Early stopping triggered in Phase 2")
            break

    model.load_state_dict(best_model_wts)
    print(f"\nTraining finished. Best validation accuracy: {best_val_acc:.4f}")

    return model, history
```

```python
import os
import cv2
import math
import tempfile
import warnings
import numpy as np
import pandas as pd
import librosa
import torch

from deepface import DeepFace
from moviepy import VideoFileClip
from transformers import AutoFeatureExtractor, AutoModelForAudioClassification

warnings.filterwarnings("ignore")

# ================= CONFIG =================
VIDEO_FOLDER = r"E:\Sohan\Data\N1"
OUTPUT_CSV = r"E:\Sohan\Data\labels.csv"

FRAME_SAMPLE_EVERY = 10     # analyze every 10th frame
AUDIO_SR = 16000
MAX_AUDIO_SEC = 6           # analyze up to first 6 sec of audio
VIDEO_WEIGHT = 0.6
AUDIO_WEIGHT = 0.4

# Hugging Face SER model
AUDIO_MODEL_NAME = "ehcalabres/wav2vec2-lg-xlsr-en-speech-emotion-recognition"
# ==========================================


# ---------------------------
# 1. Load audio SER model
# ---------------------------
print("Loading audio emotion model...")
feature_extractor = AutoFeatureExtractor.from_pretrained(AUDIO_MODEL_NAME)
audio_model = AutoModelForAudioClassification.from_pretrained(AUDIO_MODEL_NAME)

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
audio_model.to(device)
audio_model.eval()

id2label = audio_model.config.id2label


# ---------------------------
# 2. Label normalization
# ---------------------------
def normalize_label(label):
    label = str(label).strip().lower()

    mapping = {
        "angry": "angry",
        "anger": "angry",
        "sad": "sad",
        "happy": "happy",
        "joy": "happy",
        "neutral": "neutral",
        "fear": "fear",
        "fearful": "fear",
        "surprise": "surprise",
        "surprised": "surprise",
        "disgust": "disgust",
        "disgusted": "disgust",
        "calm": "neutral"
    }
    return mapping.get(label, "neutral")


EMOTIONS = ["happy", "sad", "angry", "neutral", "fear", "surprise", "disgust"]


# ---------------------------
# 3. Video emotion with DeepFace
# ---------------------------
def get_video_emotion(video_path, frame_sample_every=10):
    cap = cv2.VideoCapture(video_path)

    votes = []
    frame_count = 0

    while True:
        ret, frame = cap.read()
        if not ret:
            break

        if frame_count % frame_sample_every == 0:
            try:
                result = DeepFace.analyze(
                    img_path=frame,
                    actions=["emotion"],
                    enforce_detection=False,
                    silent=True
                )

                # DeepFace may return list or dict depending on version/setup
                if isinstance(result, list):
                    dominant = result[0].get("dominant_emotion", "neutral")
                else:
                    dominant = result.get("dominant_emotion", "neutral")

                votes.append(normalize_label(dominant))

            except Exception:
                pass

        frame_count += 1

    cap.release()

    if len(votes) == 0:
        return "neutral", {e: 0.0 for e in EMOTIONS}

    counts = {e: votes.count(e) for e in EMOTIONS}
    total = sum(counts.values())
    probs = {e: counts[e] / total for e in EMOTIONS}
    final_label = max(probs, key=probs.get)

    return final_label, probs


# ---------------------------
# 4. Extract temporary audio
# ---------------------------
def extract_audio_temp(video_path, sr=16000):
    tmp_wav = tempfile.NamedTemporaryFile(suffix=".wav", delete=False)
    tmp_wav_path = tmp_wav.name
    tmp_wav.close()

    clip = VideoFileClip(video_path)

    if clip.audio is None:
        clip.close()
        return None

    clip.audio.write_audiofile(
        tmp_wav_path,
        fps=sr,
        nbytes=2,
        codec="pcm_s16le",
        logger=None
    )
    clip.close()
    return tmp_wav_path


# ---------------------------
# 5. Audio emotion with Wav2Vec2 SER
# ---------------------------
def get_audio_emotion(video_path, sr=16000, max_audio_sec=6):
    tmp_audio = extract_audio_temp(video_path, sr=sr)

    if tmp_audio is None:
        return "neutral", {e: 0.0 for e in EMOTIONS}

    try:
        waveform, _ = librosa.load(tmp_audio, sr=sr, mono=True)

        max_len = sr * max_audio_sec
        if len(waveform) > max_len:
            waveform = waveform[:max_len]

        inputs = feature_extractor(
            waveform,
            sampling_rate=sr,
            return_tensors="pt",
            padding=True
        )

        inputs = {k: v.to(device) for k, v in inputs.items()}

        with torch.no_grad():
            logits = audio_model(**inputs).logits
            probs = torch.softmax(logits, dim=-1).squeeze(0).cpu().numpy()

        raw_probs = {}
        for idx, p in enumerate(probs):
            lbl = normalize_label(id2label[idx])
            raw_probs[lbl] = raw_probs.get(lbl, 0.0) + float(p)

        final_probs = {e: raw_probs.get(e, 0.0) for e in EMOTIONS}
        final_label = max(final_probs, key=final_probs.get)

        return final_label, final_probs

    except Exception:
        return "neutral", {e: 0.0 for e in EMOTIONS}

    finally:
        if os.path.exists(tmp_audio):
            os.remove(tmp_audio)


# ---------------------------
# 6. Fuse video + audio predictions
# ---------------------------
def fuse_predictions(video_probs, audio_probs, video_weight=0.6, audio_weight=0.4):
    fused = {}
    for e in EMOTIONS:
        fused[e] = video_weight * video_probs.get(e, 0.0) + audio_weight * audio_probs.get(e, 0.0)

    label = max(fused, key=fused.get)
    confidence = fused[label]
    return label, confidence, fused


# ---------------------------
# 7. Main auto-labeling loop
# ---------------------------
def auto_label_videos_multimodal(video_folder, output_csv):
    rows = []

    videos = [
        f for f in os.listdir(video_folder)
        if f.lower().endswith((".mp4", ".avi", ".mov", ".mkv", ".flv", ".wmv"))
    ]

    print(f"Found {len(videos)} videos")

    for i, video_file in enumerate(videos, 1):
        video_path = os.path.join(video_folder, video_file)
        base_name = os.path.splitext(video_file)[0]

        print(f"\n[{i}/{len(videos)}] Processing: {video_file}")

        video_label, video_probs = get_video_emotion(video_path, frame_sample_every=FRAME_SAMPLE_EVERY)
        audio_label, audio_probs = get_audio_emotion(video_path, sr=AUDIO_SR, max_audio_sec=MAX_AUDIO_SEC)

        final_label, confidence, fused_probs = fuse_predictions(
            video_probs,
            audio_probs,
            video_weight=VIDEO_WEIGHT,
            audio_weight=AUDIO_WEIGHT
        )

        print(f"  video label : {video_label}")
        print(f"  audio label : {audio_label}")
        print(f"  final label : {final_label} ({confidence:.3f})")

        rows.append({
            "file_name": base_name,
            "label": final_label,
            "confidence": round(confidence, 4),
            "video_label": video_label,
            "audio_label": audio_label
        })

    df = pd.DataFrame(rows)
    df.to_csv(output_csv, index=False)

    print(f"\nSaved auto labels to: {output_csv}")
    return df


# ---------------------------
# 8. Run
# ---------------------------
df = auto_label_videos_multimodal(VIDEO_FOLDER, OUTPUT_CSV)
print(df.head())
```

```python
def save_waveforms_from_audio_folder(audio_folder, waveform_output_folder, sample_rate=16000, duration=3.0):
    os.makedirs(waveform_output_folder, exist_ok=True)
    target_length = int(sample_rate * duration)

    supported_formats = (".wav", ".mp3", ".flac", ".ogg", ".m4a")

    for file_name in os.listdir(audio_folder):
        if file_name.lower().endswith(supported_formats):
            audio_path = os.path.join(audio_folder, file_name)
            base_name = os.path.splitext(file_name)[0]

            try:
                waveform, sr = librosa.load(audio_path, sr=sample_rate, mono=True)

                if len(waveform) < target_length:
                    pad_width = target_length - len(waveform)
                    waveform = np.pad(waveform, (0, pad_width), mode="constant")
                elif len(waveform) > target_length:
                    waveform = waveform[:target_length]

                save_path = os.path.join(waveform_output_folder, f"{base_name}.npy")
                np.save(save_path, waveform.astype(np.float32))

                print(f"Saved waveform: {save_path}")

            except Exception as e:
                print(f"Error processing {audio_path}: {e}")
```

```python
import os
import numpy as np
import pandas as pd
import torch
from torch.utils.data import Dataset


class MultimodalEmotionDataset(Dataset):
    def __init__(
        self,
        labels_csv,
        frames_dir,
        mel_dir,
        waveform_dir,
        label_map=None,
        num_frames=16,
        frame_size=(224, 224),
        n_mels=128,
        waveform_length=48000
    ):
        self.frames_dir = frames_dir
        self.mel_dir = mel_dir
        self.waveform_dir = waveform_dir
        self.num_frames = num_frames
        self.frame_size = frame_size
        self.n_mels = n_mels
        self.waveform_length = waveform_length

        self.label_map = label_map or {
            "happy": 0,
            "sad": 1,
            "angry": 2,
            "neutral": 3,
            "fear": 4,
            "surprise": 5,
            "disgust": 6
        }

        if not os.path.exists(labels_csv):
            raise FileNotFoundError(f"labels.csv not found: {labels_csv}")
        if not os.path.exists(frames_dir):
            raise FileNotFoundError(f"processed_frames folder not found: {frames_dir}")
        if not os.path.exists(mel_dir):
            raise FileNotFoundError(f"processed_audio folder not found: {mel_dir}")
        if not os.path.exists(waveform_dir):
            raise FileNotFoundError(f"waveforms folder not found: {waveform_dir}")

        self.data = pd.read_csv(labels_csv)

        if "file_name" not in self.data.columns:
            raise ValueError("CSV must contain a 'file_name' column")

        if "label" in self.data.columns:
            self.label_col = "label"
        elif "emotion" in self.data.columns:
            self.label_col = "emotion"
        else:
            raise ValueError("CSV must contain either 'label' or 'emotion' column")

        self.data["file_name"] = self.data["file_name"].apply(self._clean_file_name)
        self.data[self.label_col] = (
            self.data[self.label_col]
            .astype(str)
            .str.strip()
            .str.lower()
        )

        self.data = self.data[
            (self.data["file_name"] != "") &
            (self.data[self.label_col] != "")
        ].reset_index(drop=True)

        self.data = self._validate_rows()

        if len(self.data) == 0:
            raise ValueError("No valid samples found after validation.")

        print(f"Final valid dataset size: {len(self.data)}")

    def _clean_file_name(self, x):
        if pd.isna(x):
            return ""

        try:
            f = float(x)
            if f.is_integer():
                return str(int(f))
        except:
            pass

        x = str(x).strip()

        for ext in [".mp4", ".avi", ".mov", ".mkv", ".wav"]:
            if x.lower().endswith(ext):
                x = x[:-len(ext)]

        if x.endswith(".0"):
            try:
                f = float(x)
                if f.is_integer():
                    x = str(int(f))
            except:
                pass

        return x

    def _validate_rows(self):
        valid_rows = []

        missing_frames = 0
        missing_mel = 0
        missing_waveform = 0
        bad_label = 0

        for _, row in self.data.iterrows():
            file_name = row["file_name"]
            label = row[self.label_col]

            frame_path = os.path.join(self.frames_dir, f"{file_name}.npy")
            mel_path = os.path.join(self.mel_dir, f"{file_name}_mel.npy")
            waveform_path = os.path.join(self.waveform_dir, f"{file_name}.npy")

            if not os.path.exists(frame_path):
                missing_frames += 1
                continue
            if not os.path.exists(mel_path):
                missing_mel += 1
                continue
            if not os.path.exists(waveform_path):
                missing_waveform += 1
                continue
            if label not in self.label_map:
                bad_label += 1
                continue

            valid_rows.append({
                "file_name": file_name,
                self.label_col: label
            })

        print("Validation summary:")
        print("  valid rows       :", len(valid_rows))
        print("  missing frames   :", missing_frames)
        print("  missing mel      :", missing_mel)
        print("  missing waveform :", missing_waveform)
        print("  bad labels       :", bad_label)

        return pd.DataFrame(valid_rows)

    def __len__(self):
        return len(self.data)

    def _fix_frames(self, frames):
        if frames.ndim != 4:
            raise ValueError(f"Frames must be 4D, got {frames.shape}")

        if frames.shape[-1] == 3:
            frames = np.transpose(frames, (0, 3, 1, 2))

        T = frames.shape[0]

        if T == 0:
            return np.zeros(
                (self.num_frames, 3, self.frame_size[0], self.frame_size[1]),
                dtype=np.float32
            )

        if T < self.num_frames:
            pad_count = self.num_frames - T
            last_frame = frames[-1:]
            pad_frames = np.repeat(last_frame, pad_count, axis=0)
            frames = np.concatenate([frames, pad_frames], axis=0)
        elif T > self.num_frames:
            idx = np.linspace(0, T - 1, self.num_frames).astype(int)
            frames = frames[idx]

        return frames.astype(np.float32)

    def _fix_mel(self, mel):
        if mel.ndim != 2:
            raise ValueError(f"Mel spectrogram must be 2D, got {mel.shape}")

        return np.expand_dims(mel.astype(np.float32), axis=0)

    def _fix_waveform(self, waveform):
        waveform = np.squeeze(waveform)

        if waveform.ndim != 1:
            raise ValueError(f"Waveform must be 1D, got {waveform.shape}")

        n = len(waveform)

        if n < self.waveform_length:
            pad_width = self.waveform_length - n
            waveform = np.pad(waveform, (0, pad_width), mode="constant")
        elif n > self.waveform_length:
            waveform = waveform[:self.waveform_length]

        return waveform.astype(np.float32)

    def __getitem__(self, idx):
        row = self.data.iloc[idx]
        file_name = row["file_name"]
        label_name = row[self.label_col]

        frame_path = os.path.join(self.frames_dir, f"{file_name}.npy")
        mel_path = os.path.join(self.mel_dir, f"{file_name}_mel.npy")
        waveform_path = os.path.join(self.waveform_dir, f"{file_name}.npy")

        frames = np.load(frame_path)
        mel = np.load(mel_path)
        waveform = np.load(waveform_path)

        frames = self._fix_frames(frames)
        mel = self._fix_mel(mel)
        waveform = self._fix_waveform(waveform)

        label = self.label_map[label_name]

        return {
            "video": torch.tensor(frames, dtype=torch.float32),
            "mel": torch.tensor(mel, dtype=torch.float32),
            "waveform": torch.tensor(waveform, dtype=torch.float32),
            "label": torch.tensor(label, dtype=torch.long)
        }
```

```python
import copy
import torch
import torch.nn as nn
from sklearn.metrics import accuracy_score


def evaluate_model(model, data_loader, device, criterion=None):
    model.eval()
    total_loss = 0.0
    all_preds = []
    all_labels = []

    with torch.no_grad():
        for batch in data_loader:
            video = batch["video"].to(device)
            mel = batch["mel"].to(device)
            waveform = batch["waveform"].to(device)
            labels = batch["label"].to(device)

            outputs = model(video, mel, waveform)

            if criterion is not None:
                loss = criterion(outputs, labels)
                total_loss += loss.item()

            preds = torch.argmax(outputs, dim=1)
            all_preds.extend(preds.cpu().numpy())
            all_labels.extend(labels.cpu().numpy())

    avg_loss = total_loss / len(data_loader) if criterion is not None and len(data_loader) > 0 else 0.0
    acc = accuracy_score(all_labels, all_preds) if len(all_labels) > 0 else 0.0
    return avg_loss, acc


def train_one_epoch(model, data_loader, optimizer, criterion, device):
    model.train()
    total_loss = 0.0
    all_preds = []
    all_labels = []

    for batch in data_loader:
        video = batch["video"].to(device)
        mel = batch["mel"].to(device)
        waveform = batch["waveform"].to(device)
        labels = batch["label"].to(device)

        optimizer.zero_grad()

        outputs = model(video, mel, waveform)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()

        total_loss += loss.item()

        preds = torch.argmax(outputs, dim=1)
        all_preds.extend(preds.detach().cpu().numpy())
        all_labels.extend(labels.detach().cpu().numpy())

    avg_loss = total_loss / len(data_loader) if len(data_loader) > 0 else 0.0
    acc = accuracy_score(all_labels, all_preds) if len(all_labels) > 0 else 0.0
    return avg_loss, acc


def train_phase(
    model,
    train_loader,
    val_loader,
    optimizer,
    criterion,
    device,
    epochs,
    patience,
    save_path,
    phase_name="phase"
):
    history = {
        f"{phase_name}_train_loss": [],
        f"{phase_name}_train_acc": [],
        f"{phase_name}_val_loss": [],
        f"{phase_name}_val_acc": []
    }

    best_val_loss = float("inf")
    best_state = copy.deepcopy(model.state_dict())
    wait = 0

    for epoch in range(epochs):
        train_loss, train_acc = train_one_epoch(model, train_loader, optimizer, criterion, device)
        val_loss, val_acc = evaluate_model(model, val_loader, device, criterion)

        history[f"{phase_name}_train_loss"].append(train_loss)
        history[f"{phase_name}_train_acc"].append(train_acc)
        history[f"{phase_name}_val_loss"].append(val_loss)
        history[f"{phase_name}_val_acc"].append(val_acc)

        print(
            f"{phase_name} Epoch [{epoch+1}/{epochs}] | "
            f"Train Loss: {train_loss:.4f} | Train Acc: {train_acc:.4f} | "
            f"Val Loss: {val_loss:.4f} | Val Acc: {val_acc:.4f}"
        )

        if val_loss < best_val_loss:
            best_val_loss = val_loss
            best_state = copy.deepcopy(model.state_dict())
            torch.save(best_state, save_path)
            wait = 0
        else:
            wait += 1
            if wait >= patience:
                print(f"Early stopping in {phase_name}")
                break

    model.load_state_dict(best_state)
    return model, history


def train_freeze_then_finetune_improved(
    model,
    train_loader,
    val_loader,
    train_dataset,
    num_classes,
    device,
    setup_phase_1_fn,
    setup_phase_2_fn,
    phase1_epochs=20,
    phase2_epochs=10,
    phase1_lr=1e-4,
    phase2_lr=1e-5,
    patience=5,
    save_path="best_multimodal_model_improved.pth"
):
    model = model.to(device)

    criterion = nn.CrossEntropyLoss()

    full_history = {}

    # Phase 1
    print("\nStarting Phase 1")
    model = setup_phase_1_fn(model)

    optimizer_phase1 = torch.optim.Adam(
        filter(lambda p: p.requires_grad, model.parameters()),
        lr=phase1_lr
    )

    phase1_save_path = save_path.replace(".pth", "_phase1.pth")
    model, history1 = train_phase(
        model=model,
        train_loader=train_loader,
        val_loader=val_loader,
        optimizer=optimizer_phase1,
        criterion=criterion,
        device=device,
        epochs=phase1_epochs,
        patience=patience,
        save_path=phase1_save_path,
        phase_name="phase1"
    )
    full_history.update(history1)

    # Phase 2
    print("\nStarting Phase 2")
    model = setup_phase_2_fn(model)

    optimizer_phase2 = torch.optim.Adam(
        filter(lambda p: p.requires_grad, model.parameters()),
        lr=phase2_lr
    )

    model, history2 = train_phase(
        model=model,
        train_loader=train_loader,
        val_loader=val_loader,
        optimizer=optimizer_phase2,
        criterion=criterion,
        device=device,
        epochs=phase2_epochs,
        patience=patience,
        save_path=save_path,
        phase_name="phase2"
    )
    full_history.update(history2)

    return model, full_history
```

```python
import torch
import torch.nn as nn
import torch.nn.functional as F


class MultimodalFusionModel(nn.Module):
    def __init__(self, num_classes=7):
        super().__init__()

        # ===== VIDEO BRANCH =====
        self.video_cnn = nn.Sequential(
            nn.Conv2d(3, 16, 3, padding=1),
            nn.ReLU(),
            nn.MaxPool2d(2),

            nn.Conv2d(16, 32, 3, padding=1),
            nn.ReLU(),
            nn.MaxPool2d(2)
        )

        self.video_fc = nn.Linear(32 * 56 * 56, 128)

        # ===== MEL BRANCH =====
        self.mel_cnn = nn.Sequential(
            nn.Conv2d(1, 16, 3, padding=1),
            nn.ReLU(),
            nn.MaxPool2d(2),

            nn.Conv2d(16, 32, 3, padding=1),
            nn.ReLU(),
            nn.AdaptiveAvgPool2d((8, 8))
        )

        self.mel_fc = nn.Linear(32 * 8 * 8, 128)

        # ===== WAVEFORM BRANCH =====
        self.wave_fc = nn.Sequential(
            nn.Linear(48000, 256),
            nn.ReLU(),
            nn.Linear(256, 128)
        )

        # ===== FUSION =====
        self.classifier = nn.Sequential(
            nn.Linear(128 * 3, 128),
            nn.ReLU(),
            nn.Dropout(0.3),
            nn.Linear(128, num_classes)
        )

    def forward(self, video, mel, waveform):
        # video: (B, T, C, H, W)
        B, T, C, H, W = video.shape

        # take mean over frames
        video = video.mean(dim=1)  # (B, C, H, W)

        v = self.video_cnn(video)
        v = v.view(B, -1)
        v = self.video_fc(v)

        m = self.mel_cnn(mel)
        m = m.view(B, -1)
        m = self.mel_fc(m)

        w = self.wave_fc(waveform)

        x = torch.cat([v, m, w], dim=1)

        out = self.classifier(x)
        return out
```

```python
import torch
import torch.nn as nn


def freeze_module(module):
    for param in module.parameters():
        param.requires_grad = False


def unfreeze_module(module):
    for param in module.parameters():
        param.requires_grad = True


def unfreeze_last_resnet_block(video_extractor):
    """
    Unfreeze the last CNN block of ResNet18 inside video_extractor.cnn
    """
    if hasattr(video_extractor, "cnn") and len(video_extractor.cnn) > 7:
        for param in video_extractor.cnn[7].parameters():
            param.requires_grad = True


def unfreeze_top_wav2vec_layers(audio_extractor, num_layers=2):
    """
    Unfreeze top wav2vec transformer layers if available
    """
    if not hasattr(audio_extractor, "wav_branch"):
        return
    if not hasattr(audio_extractor.wav_branch, "wav2vec"):
        return

    wav2vec_model = audio_extractor.wav_branch.wav2vec

    # freeze all first
    freeze_module(wav2vec_model)

    try:
        layers = wav2vec_model.encoder.transformer.layers
        for layer in layers[-num_layers:]:
            for param in layer.parameters():
                param.requires_grad = True
    except Exception:
        pass
```

```python
def setup_phase_1(model):
    """
    Phase 1:
    Freeze pretrained backbones
    Train only task-specific layers
    """

    # freeze video CNN
    if hasattr(model, "video_extractor") and hasattr(model.video_extractor, "cnn"):
        freeze_module(model.video_extractor.cnn)

    # freeze wav2vec
    if hasattr(model, "audio_extractor"):
        if hasattr(model.audio_extractor, "wav_branch") and hasattr(model.audio_extractor.wav_branch, "wav2vec"):
            freeze_module(model.audio_extractor.wav_branch.wav2vec)

    # keep task-specific layers trainable
    if hasattr(model, "video_extractor"):
        if hasattr(model.video_extractor, "lstm"):
            unfreeze_module(model.video_extractor.lstm)
        if hasattr(model.video_extractor, "fc"):
            unfreeze_module(model.video_extractor.fc)

    if hasattr(model, "audio_extractor"):
        if hasattr(model.audio_extractor, "mel_branch"):
            unfreeze_module(model.audio_extractor.mel_branch)
        if hasattr(model.audio_extractor, "wav_branch") and hasattr(model.audio_extractor.wav_branch, "proj"):
            unfreeze_module(model.audio_extractor.wav_branch.proj)
        if hasattr(model.audio_extractor, "fusion"):
            unfreeze_module(model.audio_extractor.fusion)

    if hasattr(model, "fusion_layer"):
        unfreeze_module(model.fusion_layer)

    if hasattr(model, "classifier"):
        unfreeze_module(model.classifier)


def setup_phase_2(model):
    """
    Phase 2:
    Fine-tune selected pretrained layers
    """

    # unfreeze last ResNet block
    if hasattr(model, "video_extractor"):
        unfreeze_last_resnet_block(model.video_extractor)

        if hasattr(model.video_extractor, "lstm"):
            unfreeze_module(model.video_extractor.lstm)
        if hasattr(model.video_extractor, "fc"):
            unfreeze_module(model.video_extractor.fc)

    # unfreeze top wav2vec layers
    if hasattr(model, "audio_extractor"):
        unfreeze_top_wav2vec_layers(model.audio_extractor, num_layers=2)

        if hasattr(model.audio_extractor, "mel_branch"):
            unfreeze_module(model.audio_extractor.mel_branch)
        if hasattr(model.audio_extractor, "wav_branch") and hasattr(model.audio_extractor.wav_branch, "proj"):
            unfreeze_module(model.audio_extractor.wav_branch.proj)
        if hasattr(model.audio_extractor, "fusion"):
            unfreeze_module(model.audio_extractor.fusion)

    if hasattr(model, "fusion_layer"):
        unfreeze_module(model.fusion_layer)

    if hasattr(model, "classifier"):
        unfreeze_module(model.classifier)
```

```python
import torch
import torch.nn as nn
import torchvision.models as models


class VideoFeatureExtractor(nn.Module):
    def __init__(self, hidden_dim=256):
        super(VideoFeatureExtractor, self).__init__()

        resnet = models.resnet18(weights=models.ResNet18_Weights.DEFAULT)

        # remove classifier
        self.cnn = nn.Sequential(*list(resnet.children())[:-1])  # (B*T, 512, 1, 1)
        self.feature_dim = 512

        self.lstm = nn.LSTM(
            input_size=self.feature_dim,
            hidden_size=hidden_dim,
            batch_first=True
        )

        self.fc = nn.Linear(hidden_dim, hidden_dim)

    def forward(self, x):
        # x: (B, T, C, H, W)
        B, T, C, H, W = x.shape

        x = x.view(B*T, C, H, W)
        x = self.cnn(x)
        x = x.view(B, T, self.feature_dim)

        x, _ = self.lstm(x)
        x = x[:, -1, :]

        return self.fc(x)
```

```python
import torchaudio


class MelCNNBranch(nn.Module):
    def __init__(self, out_dim=256):
        super().__init__()

        self.net = nn.Sequential(
            nn.Conv2d(1, 32, 3, padding=1),
            nn.ReLU(),
            nn.MaxPool2d(2),

            nn.Conv2d(32, 64, 3, padding=1),
            nn.ReLU(),
            nn.AdaptiveAvgPool2d((1, 1))
        )

        self.fc = nn.Linear(64, out_dim)

    def forward(self, x):
        x = self.net(x)
        x = x.view(x.size(0), -1)
        return self.fc(x)


class Wav2VecBranch(nn.Module):
    def __init__(self, out_dim=256, freeze=True):
        super().__init__()

        bundle = torchaudio.pipelines.WAV2VEC2_BASE
        self.wav2vec = bundle.get_model()

        if freeze:
            for p in self.wav2vec.parameters():
                p.requires_grad = False

        self.proj = nn.Linear(768, out_dim)

    def forward(self, waveform):
        features, _ = self.wav2vec.extract_features(waveform)
        x = features[-1]          # (B, T, 768)
        x = x.mean(dim=1)         # (B, 768)
        return self.proj(x)


class HybridAudioFeatureExtractor(nn.Module):
    def __init__(self, cnn_feature_dim=256, wav2vec_feature_dim=256, fusion_hidden_dim=256, freeze_wav2vec=True):
        super().__init__()

        self.mel_branch = MelCNNBranch(out_dim=cnn_feature_dim)
        self.wav_branch = Wav2VecBranch(out_dim=wav2vec_feature_dim, freeze=freeze_wav2vec)

        self.fusion = nn.Sequential(
            nn.Linear(cnn_feature_dim + wav2vec_feature_dim, fusion_hidden_dim),
            nn.ReLU()
        )

    def forward(self, mel, waveform):
        mel_feat = self.mel_branch(mel)
        wav_feat = self.wav_branch(waveform)

        x = torch.cat([mel_feat, wav_feat], dim=1)
        return self.fusion(x)
```

```python
class FullMultimodalEmotionModel(nn.Module):
    def __init__(
        self,
        video_extractor,
        audio_extractor,
        video_feature_dim=256,
        audio_feature_dim=256,
        fusion_dim=256,
        num_classes=7
    ):
        super().__init__()

        self.video_extractor = video_extractor
        self.audio_extractor = audio_extractor

        self.fusion_layer = nn.Sequential(
            nn.Linear(video_feature_dim + audio_feature_dim, fusion_dim),
            nn.ReLU()
        )

        self.classifier = nn.Linear(fusion_dim, num_classes)

    def forward(self, video, mel, waveform):
        v = self.video_extractor(video)
        a = self.audio_extractor(mel, waveform)

        x = torch.cat([v, a], dim=1)
        x = self.fusion_layer(x)

        out = self.classifier(x)
        return out, x
```

```python
class MultimodalEmotionDataset(Dataset):
    def __init__(
        self,
        labels_csv,
        frames_dir,
        mel_dir,
        waveform_dir,
        label_map=None,
        num_frames=16,
        frame_size=(224, 224),
        waveform_length=48000
    ):
        self.frames_dir = frames_dir
        self.mel_dir = mel_dir
        self.waveform_dir = waveform_dir
        self.num_frames = num_frames
        self.frame_size = frame_size
        self.waveform_length = waveform_length

        self.label_map = label_map or {
            "happy": 0,
            "sad": 1,
            "angry": 2,
            "neutral": 3,
            "fear": 4,
            "surprise": 5,
            "disgust": 6
        }

        if not os.path.exists(labels_csv):
            raise FileNotFoundError(f"labels.csv not found: {labels_csv}")
        if not os.path.exists(frames_dir):
            raise FileNotFoundError(f"processed_frames folder not found: {frames_dir}")
        if not os.path.exists(mel_dir):
            raise FileNotFoundError(f"processed_audio folder not found: {mel_dir}")
        if not os.path.exists(waveform_dir):
            raise FileNotFoundError(f"waveforms folder not found: {waveform_dir}")

        self.data = pd.read_csv(labels_csv)

        if "file_name" not in self.data.columns:
            raise ValueError("CSV must contain a 'file_name' column")

        if "label" in self.data.columns:
            self.label_col = "label"
        elif "emotion" in self.data.columns:
            self.label_col = "emotion"
        else:
            raise ValueError("CSV must contain either 'label' or 'emotion' column")

        self.data["file_name"] = self.data["file_name"].apply(self._clean_file_name)
        self.data[self.label_col] = (
            self.data[self.label_col]
            .astype(str)
            .str.strip()
            .str.lower()
        )

        self.data = self.data[
            (self.data["file_name"] != "") &
            (self.data[self.label_col] != "")
        ].reset_index(drop=True)

        self.data = self._validate_rows()

        if len(self.data) == 0:
            raise ValueError("No valid samples found after validation.")

        print(f"Final valid dataset size: {len(self.data)}")

    def _clean_file_name(self, x):
        if pd.isna(x):
            return ""

        try:
            f = float(x)
            if f.is_integer():
                return str(int(f))
        except:
            pass

        x = str(x).strip()

        for ext in [".mp4", ".avi", ".mov", ".mkv", ".wav"]:
            if x.lower().endswith(ext):
                x = x[:-len(ext)]

        if x.endswith(".0"):
            try:
                f = float(x)
                if f.is_integer():
                    x = str(int(f))
            except:
                pass

        return x

    def _validate_rows(self):
        valid_rows = []

        missing_frames = 0
        missing_mel = 0
        missing_waveform = 0
        bad_label = 0

        for _, row in self.data.iterrows():
            file_name = row["file_name"]
            label = row[self.label_col]

            frame_path = os.path.join(self.frames_dir, f"{file_name}.npy")
            mel_path = os.path.join(self.mel_dir, f"{file_name}_mel.npy")
            waveform_path = os.path.join(self.waveform_dir, f"{file_name}.npy")

            if not os.path.exists(frame_path):
                missing_frames += 1
                continue
            if not os.path.exists(mel_path):
                missing_mel += 1
                continue
            if not os.path.exists(waveform_path):
                missing_waveform += 1
                continue
            if label not in self.label_map:
                bad_label += 1
                continue

            valid_rows.append({
                "file_name": file_name,
                self.label_col: label
            })

        print("Validation summary:")
        print("  valid rows       :", len(valid_rows))
        print("  missing frames   :", missing_frames)
        print("  missing mel      :", missing_mel)
        print("  missing waveform :", missing_waveform)
        print("  bad labels       :", bad_label)

        return pd.DataFrame(valid_rows)

    def __len__(self):
        return len(self.data)

    def _fix_frames(self, frames):
        if frames.ndim != 4:
            raise ValueError(f"Frames must be 4D, got {frames.shape}")

        if frames.shape[-1] == 3:
            frames = np.transpose(frames, (0, 3, 1, 2))

        T = frames.shape[0]

        if T == 0:
            return np.zeros(
                (self.num_frames, 3, self.frame_size[0], self.frame_size[1]),
                dtype=np.float32
            )

        if T < self.num_frames:
            pad_count = self.num_frames - T
            last_frame = frames[-1:]
            pad_frames = np.repeat(last_frame, pad_count, axis=0)
            frames = np.concatenate([frames, pad_frames], axis=0)
        elif T > self.num_frames:
            idx = np.linspace(0, T - 1, self.num_frames).astype(int)
            frames = frames[idx]

        return frames.astype(np.float32)

    def _fix_mel(self, mel):
        if mel.ndim != 2:
            raise ValueError(f"Mel spectrogram must be 2D, got {mel.shape}")

        return np.expand_dims(mel.astype(np.float32), axis=0)

    def _fix_waveform(self, waveform):
        waveform = np.squeeze(waveform)

        if waveform.ndim != 1:
            raise ValueError(f"Waveform must be 1D, got {waveform.shape}")

        n = len(waveform)

        if n < self.waveform_length:
            pad_width = self.waveform_length - n
            waveform = np.pad(waveform, (0, pad_width), mode="constant")
        elif n > self.waveform_length:
            waveform = waveform[:self.waveform_length]

        return waveform.astype(np.float32)

    def __getitem__(self, idx):
        row = self.data.iloc[idx]
        file_name = row["file_name"]
        label_name = row[self.label_col]

        frame_path = os.path.join(self.frames_dir, f"{file_name}.npy")
        mel_path = os.path.join(self.mel_dir, f"{file_name}_mel.npy")
        waveform_path = os.path.join(self.waveform_dir, f"{file_name}.npy")

        frames = np.load(frame_path)
        mel = np.load(mel_path)
        waveform = np.load(waveform_path)

        frames = self._fix_frames(frames)
        mel = self._fix_mel(mel)
        waveform = self._fix_waveform(waveform)

        if frames.max() > 1:
            frames = frames / 255.0

        label = self.label_map[label_name]

        return {
            "video": torch.tensor(frames, dtype=torch.float32),
            "mel": torch.tensor(mel, dtype=torch.float32),
            "waveform": torch.tensor(waveform, dtype=torch.float32),
            "label": torch.tensor(label, dtype=torch.long)
        }
```

```python
class MultimodalFusionModel(nn.Module):
    def __init__(self, num_classes=7):
        super().__init__()

        # video branch
        self.video_cnn = nn.Sequential(
            nn.Conv2d(3, 16, kernel_size=3, padding=1),
            nn.BatchNorm2d(16),
            nn.ReLU(),
            nn.MaxPool2d(2),

            nn.Conv2d(16, 32, kernel_size=3, padding=1),
            nn.BatchNorm2d(32),
            nn.ReLU(),
            nn.MaxPool2d(2),

            nn.Conv2d(32, 64, kernel_size=3, padding=1),
            nn.BatchNorm2d(64),
            nn.ReLU(),
            nn.AdaptiveAvgPool2d((7, 7))
        )

        self.video_fc = nn.Sequential(
            nn.Linear(64 * 7 * 7, 128),
            nn.ReLU(),
            nn.Dropout(0.3)
        )

        # mel branch
        self.mel_cnn = nn.Sequential(
            nn.Conv2d(1, 16, kernel_size=3, padding=1),
            nn.BatchNorm2d(16),
            nn.ReLU(),
            nn.MaxPool2d(2),

            nn.Conv2d(16, 32, kernel_size=3, padding=1),
            nn.BatchNorm2d(32),
            nn.ReLU(),
            nn.AdaptiveAvgPool2d((8, 8))
        )

        self.mel_fc = nn.Sequential(
            nn.Linear(32 * 8 * 8, 128),
            nn.ReLU(),
            nn.Dropout(0.3)
        )

        # waveform branch
        self.wave_fc = nn.Sequential(
            nn.Linear(48000, 512),
            nn.ReLU(),
            nn.Dropout(0.3),
            nn.Linear(512, 128),
            nn.ReLU()
        )

        # gating
        self.gate = nn.Sequential(
            nn.Linear(128 * 3, 128),
            nn.ReLU(),
            nn.Linear(128, 3),
            nn.Softmax(dim=1)
        )

        # final classifier
        self.classifier = nn.Sequential(
            nn.Linear(128 * 3, 256),
            nn.ReLU(),
            nn.Dropout(0.4),
            nn.Linear(256, num_classes)
        )

    def forward(self, video, mel, waveform):
        # video: (B, T, C, H, W)
        B, T, C, H, W = video.shape

        # average over time
        video = video.mean(dim=1)

        v = self.video_cnn(video)
        v = v.view(B, -1)
        v = self.video_fc(v)

        m = self.mel_cnn(mel)
        m = m.view(B, -1)
        m = self.mel_fc(m)

        w = self.wave_fc(waveform)

        fusion = torch.cat([v, m, w], dim=1)
        gates = self.gate(fusion)

        gv = gates[:, 0:1]
        gm = gates[:, 1:2]
        gw = gates[:, 2:3]

        v = v * gv
        m = m * gm
        w = w * gw

        x = torch.cat([v, m, w], dim=1)
        return self.classifier(x)
```

```python
model = MultimodalFusionModel(num_classes=7).to(device)
print(model.__class__.__name__)
```

```python
def setup_phase_1(model):
    for param in model.parameters():
        param.requires_grad = False

    modules_to_train = [
        model.video_fc,
        model.mel_fc,
        model.wave_fc,
        model.gate,
        model.classifier
    ]

    for module in modules_to_train:
        for param in module.parameters():
            param.requires_grad = True

    return model


def setup_phase_2(model):
    for param in model.parameters():
        param.requires_grad = True

    return model
```

```python
def evaluate_model(model, data_loader, device, criterion=None):
    model.eval()
    total_loss = 0.0
    correct = 0
    total = 0

    with torch.no_grad():
        for batch in data_loader:
            video = batch["video"].to(device)
            mel = batch["mel"].to(device)
            waveform = batch["waveform"].to(device)
            labels = batch["label"].to(device)

            outputs = model(video, mel, waveform)

            if criterion is not None:
                loss = criterion(outputs, labels)
                total_loss += loss.item()

            preds = torch.argmax(outputs, dim=1)
            correct += (preds == labels).sum().item()
            total += labels.size(0)

    avg_loss = total_loss / len(data_loader) if criterion is not None and len(data_loader) > 0 else 0.0
    acc = correct / total if total > 0 else 0.0
    return avg_loss, acc


def train_one_epoch(model, data_loader, optimizer, criterion, device):
    model.train()
    total_loss = 0.0
    correct = 0
    total = 0

    for batch in data_loader:
        video = batch["video"].to(device)
        mel = batch["mel"].to(device)
        waveform = batch["waveform"].to(device)
        labels = batch["label"].to(device)

        optimizer.zero_grad()
        outputs = model(video, mel, waveform)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()

        total_loss += loss.item()
        preds = torch.argmax(outputs, dim=1)
        correct += (preds == labels).sum().item()
        total += labels.size(0)

    avg_loss = total_loss / len(data_loader) if len(data_loader) > 0 else 0.0
    acc = correct / total if total > 0 else 0.0
    return avg_loss, acc


def train_phase(
    model,
    train_loader,
    val_loader,
    optimizer,
    criterion,
    device,
    epochs,
    patience,
    save_path,
    phase_name="phase"
):
    history = {
        f"{phase_name}_train_loss": [],
        f"{phase_name}_train_acc": [],
        f"{phase_name}_val_loss": [],
        f"{phase_name}_val_acc": []
    }

    best_val_loss = float("inf")
    best_state = copy.deepcopy(model.state_dict())
    wait = 0

    for epoch in range(epochs):
        train_loss, train_acc = train_one_epoch(model, train_loader, optimizer, criterion, device)
        val_loss, val_acc = evaluate_model(model, val_loader, device, criterion)

        history[f"{phase_name}_train_loss"].append(train_loss)
        history[f"{phase_name}_train_acc"].append(train_acc)
        history[f"{phase_name}_val_loss"].append(val_loss)
        history[f"{phase_name}_val_acc"].append(val_acc)

        print(
            f"{phase_name} Epoch [{epoch+1}/{epochs}] | "
            f"Train Loss: {train_loss:.4f} | Train Acc: {train_acc:.4f} | "
            f"Val Loss: {val_loss:.4f} | Val Acc: {val_acc:.4f}"
        )

        if val_loss < best_val_loss:
            best_val_loss = val_loss
            best_state = copy.deepcopy(model.state_dict())
            torch.save(best_state, save_path)
            wait = 0
        else:
            wait += 1
            if wait >= patience:
                print(f"Early stopping in {phase_name}")
                break

    model.load_state_dict(best_state)
    return model, history


def train_freeze_then_finetune_improved(
    model,
    train_loader,
    val_loader,
    train_dataset,
    num_classes,
    device,
    setup_phase_1_fn,
    setup_phase_2_fn,
    phase1_epochs=10,
    phase2_epochs=5,
    phase1_lr=1e-4,
    phase2_lr=1e-5,
    patience=3,
    save_path="best_multimodal_model_improved.pth"
):
    model = model.to(device)
    criterion = nn.CrossEntropyLoss()
    full_history = {}

    print("\nStarting Phase 1")
    model = setup_phase_1_fn(model)

    if model is None:
        raise ValueError("setup_phase_1_fn returned None. It must end with: return model")

    optimizer_phase1 = torch.optim.Adam(
        filter(lambda p: p.requires_grad, model.parameters()),
        lr=phase1_lr
    )

    phase1_save_path = save_path.replace(".pth", "_phase1.pth")
    model, history1 = train_phase(
        model=model,
        train_loader=train_loader,
        val_loader=val_loader,
        optimizer=optimizer_phase1,
        criterion=criterion,
        device=device,
        epochs=phase1_epochs,
        patience=patience,
        save_path=phase1_save_path,
        phase_name="phase1"
    )
    full_history.update(history1)

    print("\nStarting Phase 2")
    model = setup_phase_2_fn(model)

    if model is None:
        raise ValueError("setup_phase_2_fn returned None. It must end with: return model")

    optimizer_phase2 = torch.optim.Adam(
        filter(lambda p: p.requires_grad, model.parameters()),
        lr=phase2_lr
    )

    model, history2 = train_phase(
        model=model,
        train_loader=train_loader,
        val_loader=val_loader,
        optimizer=optimizer_phase2,
        criterion=criterion,
        device=device,
        epochs=phase2_epochs,
        patience=patience,
        save_path=save_path,
        phase_name="phase2"
    )
    full_history.update(history2)

    return model, full_history
```

```python
model, history = train_freeze_then_finetune_improved(
    model=model,
    train_loader=train_loader,
    val_loader=val_loader,
    train_dataset=train_dataset,
    num_classes=7,
    device=device,
    setup_phase_1_fn=setup_phase_1,
    setup_phase_2_fn=setup_phase_2,
    phase1_epochs=10,
    phase2_epochs=5,
    phase1_lr=1e-4,
    phase2_lr=1e-5,
    patience=3,
    save_path="best_multimodal_model_improved.pth"
)
```
